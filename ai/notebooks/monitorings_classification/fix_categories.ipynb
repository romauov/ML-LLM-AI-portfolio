# Внесение корректировок в файлы с разрубами по мясу

```python
import os 
import pandas as pd
```

```python
meats_path = "../../monitorings/meat/predictor/data/categories/"
```

```python
mutton_datapath = os.path.join(meats_path, "Баранина_разрубы.csv")
bird_datapath = os.path.join(meats_path, "Птица_разрубы.csv")
pork_datapath = os.path.join(meats_path, "Свинина_разрубы.csv")
beef_datapath = os.path.join(meats_path, "Говядина_разрубы.csv")
```

```python
bird = pd.read_csv(bird_datapath, sep=';')
pork = pd.read_csv(pork_datapath, sep=';')
beef = pd.read_csv(beef_datapath, sep=';')
mutton = pd.read_csv(mutton_datapath, sep=';')
```

## Птица

**Птица_разрубы:**   
- [x] 'филе грудки', - в списках у тебя стоит просто филе, надо его заменить на филе грудки  
- [x] 'набор для студня', - добавить
- [x] 'набор для холодца', - добавить
- [X] 'грудинка' (НЕТУ!!! Есть 'грудка'), - убрать
- [x] 'набор для плова', - добавить
- [x] 'мясо механической обвалки' - добавить
- [X] 'лопатка' (НЕТУ !!!) - убрать

### Замена

```python
# создание копий датафреймов для внесения корректировок
bird_fixed = bird.copy()
beef_fixed = beef.copy()
pork_fixed = pork.copy()
mutton_fixed = mutton.copy()
```

```python
# замена филе на филе грудки
bird_fixed.loc[bird_fixed['value'] == 'филе', 'value'] = 'филе грудки'
```

```python
# замена желудки на желудок
bird_fixed.loc[bird_fixed['value'] == 'желудок', 'value'] = 'желудки'
```

```python
# замена гузки на гузка
bird_fixed.loc[bird_fixed['value'] == 'гузка', 'value'] = 'гузки'
```

### Добавление

```python
to_add_bird = ['набор для студня', 'набор для холодца', 'набор для плова', 'мясо механической обвалки', ]
```

```python
bird_fixed = pd.concat([bird_fixed, pd.DataFrame({'value': to_add_bird})], ignore_index=True)
```

```python
# сортировка по строкам
bird_fixed = bird_fixed.sort_values(by='value', key=lambda col: col.str.lower())
```

**Баранина_разрубы:** 
- [x] 'блочная' это формат фасовки (мясо, которое было заморожено в кусках весом до 25 кг). По идее надо
- [x] 'шейный отруб' – есть в списке (требуется убрать пробел)

- [x] 'лопаточно-шейный отруб' = лопаточно-шейная часть

### Замена

```python
# замена голяшка на голень
mutton_fixed.loc[mutton_fixed['value'] == 'шейный отруб ', 'value'] = 'шейный отруб'
```

```python
to_add_mutton = ['блочная', 'каре']
```

```python
mutton_fixed = mutton_fixed.sort_values(by='value', key=lambda col: col.str.lower())
```

# Сохранение

```python
# id конвертирую в Int64
# пустые поля заполняю на основе наблюдения их постоянства
bird_fixed['id'] = bird_fixed['id'].astype('Int64')
bird_fixed['parameter_id'] = 2
bird_fixed['parent_id'] = 19
bird_fixed['archive'] = 0
bird_fixed['value'] = bird_fixed['value'].str.lower()

pork_fixed['id'] = pork_fixed['id'].astype('Int64')
pork_fixed['parameter_id'] = 2
pork_fixed['parent_id'] = 12
pork_fixed['archive'] = 0
pork_fixed['value'] = pork_fixed['value'].str.lower()

beef_fixed['id'] = beef_fixed['id'].astype('Int64')
beef_fixed['parameter_id'] = 2
beef_fixed['parent_id'] = 2
beef_fixed['archive'] = 0
beef_fixed['value'] = beef_fixed['value'].str.lower()

mutton_fixed['id'] = mutton_fixed['id'].astype('Int64')
mutton_fixed['parameter_id'] = 2
mutton_fixed['parent_id'] = 1
mutton_fixed['archive'] = 0
mutton_fixed['value'] = mutton_fixed['value'].str.lower()
```

```python
def save_if_changed(original_df, fixed_df, filepath, name):
    original_sorted = original_df.sort_values(by='value', key=lambda col: col.str.lower()).reset_index(drop=True)
    fixed_sorted = fixed_df.sort_values(by='value', key=lambda col: col.str.lower()).reset_index(drop=True)
    
    if not original_sorted.equals(fixed_sorted):
        print(f"Обнаружены изменения в {name}. Сохраняю файл...")
        fixed_df.to_csv(filepath, sep=';', index=False)
        print(f"Файл {filepath} обновлен.")
    else:
        print(f"Изменения в {name} не обнаружены. Файл не перезаписывается.")

save_if_changed(bird, bird_fixed, bird_datapath, "Птица")
save_if_changed(pork, pork_fixed, pork_datapath, "Свинина") 
save_if_changed(beef, beef_fixed, beef_datapath, "Говядина")
save_if_changed(mutton, mutton_fixed, mutton_datapath, "Баранина")
```
