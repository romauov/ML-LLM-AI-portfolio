````python
import requests
import json
import pandas as pd
from sklearn.model_selection import train_test_split
````

### Подготавливаем данные

````python
data = {'text':[], 'class':[]}
with open('../dataset//dataset_COP.json', encoding = 'utf-8') as f:
    data_json = json.load(f)
    for d in data_json:
        try:
            text = d['text']
            target = d['sentiment']
            data['text'].append(text)
            data['class'].append(target)
        except:
            pass
        
with open('../dataset/dataset_COP.json', encoding = 'utf-8') as f:
    data_json = json.load(f)
    for d in data_json:
        try:
#             print(d)
            
            text = d['Text']
            target = d['sentiment']
            data['text'].append(text)
            data['class'].append(target)
        except:
            pass
        
data = pd.DataFrame(data)
data
````

````python
class_list = [
'автоответ',
'неинформативно',
'отказ',
'встречное предложение (комплиментарное)',
'встречное предложение (некомплиментарное)',
'уточнение по доставке',
'запрос прайса',
'запрос прайса и местонахождения',
'уточнение информации',
'запрос контакта',
'запрос конкретной продукции',
'запрос КП',
'запрос КП: запрос от сотрудника М16',
'запрос экспорта',
'запрос состава продукции',
'запрос фото',
'готовность к сотрудничествау: запрос документов',
'запрос образцов',
'запрос сотрудничества',
'запрос счета',
'подтверждение запроса',
'спам',
'Не могу опеределить класс',
'запрос КП: запрос прайса',
'встречное предложение (некомплементарное)',
'встречное предложение (комплементарное)',
'Пример для удаления']
````

### Очистка данных от лишней информации

````python
data = data[data['class'].isin(class_list)]

try:
    data.loc[data['class']=='встречное предложение (комплементарное)', 'class'] = 'встречное предложение (комплиментарное)'
    data.loc[data['class']=='встречное предложение (некомплементарное)', 'class'] = 'встречное предложение (некомплиментарное)'
except:
    pass

data['class'].value_counts()
````

### Возмём только те классы, количество которых больше 400

````python
data = data.loc[data['class'].isin(list(data['class'].value_counts().loc[lambda x : x >=400].index))]
data = data.reset_index(drop=True)
data['class'].value_counts()
````

````python
train, test = train_test_split(data, test_size=0.2, random_state=12345)
test, val = train_test_split(test, test_size=0.5, random_state=12345)
display(train.shape)
display(test.shape)
display(val.shape)
````
